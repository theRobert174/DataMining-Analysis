## Gráfica de dispersión

```python
def get_cmap(n, name="hsv"):
    return plt.cm.get_cmap(name, n)

def data_dispersion(labelx, labely, df):

  sexs = df['sex'].unique()
  cmap = get_cmap(len(sexs) + 1)

  plt.figure(figsize=(12, 6))

  for i, sex in enumerate(sexs):
      
      subset = df[df['sex'] == sex]

      plt.scatter(subset[labelx], subset[labely], label=sex, color=cmap(i), alpha=0.5)

  plt.legend()
  plt.title('Gráfica de Dispersión')
  plt.xlabel(labelx)
  plt.ylabel(labely)
  plt.grid(True)
  plt.show()

data_dispersion('suicides_no', 'population', df)
```
