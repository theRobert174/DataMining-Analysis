## Gráfica de dispersión

```python
def get_cmap(n, name="hsv"):
    return plt.cm.get_cmap(name, n)

def dataDispersion(label_x, label_y, df):

  sexs = df['sex'].unique()
  cmap = get_cmap(len(sexs) + 1)

  plt.figure(figsize=(12, 6))

  for i, sex in enumerate(sexs):
      
      subset = df[df['sex'] == sex]

      plt.scatter(subset[label_x], subset[label_y], label=sex, color=cmap(i), alpha=0.8)

  plt.legend()
  plt.title('Gráfica de Dispersión')
  plt.xlabel(label_x)
  plt.ylabel(label_y)
  plt.grid(color='gray', linestyle=':', linewidth=0.5)

  plt.savefig(f'../Graphs-Classification/Classification.png')
  plt.show()

dataDispersion('suicides_no', 'population', df)
```
